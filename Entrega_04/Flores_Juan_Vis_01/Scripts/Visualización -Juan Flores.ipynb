```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/content/drive/MyDrive/Entrega 4/base_de_datos_final.csv'

# Carga el archivo CSV en un marco de datos
df = pd.read_csv(file_path)

# Filtrar los datos que no tienen 'Chile' en la columna 'País'
df_filtered = df[df['País'] != 'Chile']

# Agrupar por año y contar la cantidad de canciones en cada año
songs_per_year = df_filtered.groupby('año').size()

# Asegurarse de que el año 2021 esté en el DataFrame incluso si no hay datos para ese año
if 2021 not in songs_per_year.index:
    songs_per_year[2021] = 0

# Ordenar el DataFrame por el índice de año
songs_per_year = songs_per_year.sort_index()

# Crear el diagrama de líneas
plt.figure(figsize=(10, 6))
plt.plot(songs_per_year.index, songs_per_year.values, marker='o', linestyle='--', color='b')

# Configuraciones de la gráfica
plt.title('Evolución de canciones en el top por año (excluyendo Chile)')
plt.xlabel('Año')
plt.ylabel('Número de canciones en el top')
plt.xticks([2020, 2021, 2022])  # Establece los valores del eje x
plt.grid(True)
plt.show()

```
