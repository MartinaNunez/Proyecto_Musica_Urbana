```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/content/drive/MyDrive/Entrega 4/base_de_datos_final.csv'

# Cargamos el archivo CSV en un marco de datos
df = pd.read_csv(file_path)

# Filtramos los datos que no tienen 'Chile' en la columna 'País'
df_filtered = df[df['País'] != 'Chile']

# Agrupamos por año y contamos la cantidad de canciones en cada año
songs_per_year = df_filtered.groupby('año').size()

# Consideramos el 2021 en el DataFrame incluso si no hay datos para ese año
if 2021 not in songs_per_year.index:
    songs_per_year[2021] = 0

# Ordenamos el DataFrame por el índice de año
songs_per_year = songs_per_year.sort_index()

# Creamos el diagrama de líneas
plt.figure(figsize=(10, 6))
plt.plot(songs_per_year.index, songs_per_year.values, marker='o', linestyle='--', color='purple')

# Configuraciones de la gráfica
plt.title('Evolución de canciones en el top por año (excluyendo Chile)')
plt.xlabel('Año')
plt.ylabel('Espacios utilizados en el top')
plt.xticks([2020, 2021, 2022])  # Establece los valores del eje x

# Configuración del eje y para que aumente de 1 en 1
plt.yticks(range(0, max(songs_per_year.values) + 1, 1))

plt.grid(True)
plt.show()
```
