# Visualización 1
---
Se utiliza un mapa con scatterplot. En este primer intento los círculos se sobreponen dificultando la lectura. Además, estos son pequeños en proporción al mapa.

```python
import pandas as pd
import plotly.express as px


df = pd.read_csv('/content/drive/MyDrive/Narración grafica/Entrega 5/base_de_datos_final.csv')

fig = px.scatter_geo(
    df,
    locations="País",
    locationmode="country names",
    color="streams",
    size="streams",
    hover_name="track_name",
    animation_frame="año",
    animation_group="track_name",
    title="Ultrapegados: reproducción de canciones urbanas chilenas en el mundo",
    template="plotly",
    projection="natural earth",
    size_max=50,
)

fig.show()
```

# Visualización 2
---
Se toma la decisión de agrupar toda la información de las reproducciones en un mismo círculo. Se añade el atributo hover con la lista de artistas urbanos y sus canciones.
Queda el desafío de buscar una alternativa que facilite la lectura.

```python
import pandas as pd
import plotly.express as px
# Cargar la base de datos
df = pd.read_csv('/content/drive/MyDrive/Narración grafica/Entrega 5/base_de_datos_final.csv')

# Se agrupan los datos, realizando la suma de las reproducciones de las canciones urbanas y se añaden los nombres de
# estas canciones a una lista
df_grouped = df.groupby(['año', 'País']).agg({
    'streams': 'sum',
    'track_name': ', '.join
}).reset_index()

# Se genera un mapa de puntos con georeferenciación. Se le asigna el país de donde se extrajeron los datos, agregando los datos
# de interes como el nombre, reproducciones (la que ademas le da el tamaño al circulo) y se aumenta el tamaño de los ciruculos
fig = px.scatter_geo(
    df_grouped,
    locations="País",
    locationmode="country names",
    color="streams",
    size="streams",
    hover_name="track_name",
    animation_frame="año",
    title= "Ultrapegados: reproducción de canciones urbanas chilenas en el mundo",
    template="plotly",
    projection="natural earth",
    size_max=100,
)

fig.show()
```

# Bonus
---



Se realiza un gráfico de barras con los diez artistas más escuchados. En este gráfico se consideran las canciones de cada artista independiente del país, sumando el total de reproducciones y limitándolo a diez.

```python
import plotly.express as px
import pandas as pd

# Cargar la base de datos
df = pd.read_csv('/content/drive/MyDrive/Narración grafica/Entrega 5/base_de_datos_final.csv')
top_artistas = df.groupby('artist_names')['streams'].sum().reset_index()

# Se realiza un sort para ordenar los valores de forma decendente.
top_artistas = top_artistas.sort_values(by='streams', ascending=False)

# Creación del grafico de barras, eje x "nombre de los artistas", eje y "cantidad de reproducciones"
fig = px.bar(top_artistas, x='artist_names', y='streams', title='Top de Artistas Más Escuchados')
fig.show()
```
