# Unir los archivos
---


Una vez los archivos se encuentran solo con las filas necesarias, se procede a unir las bases de datos en un solo archivo csv.

```python
import os # Importa la libreria os, que trae consigo modulos para manejar el sistema operativo.
import pandas as pd #Importa la liberia Pandas para poder utilizar sus funciones en el proyecto.


ruta_archivo = '/content/drive/MyDrive/Narración grafica/Top 2020 /Resultados/' #Se crea una variable con la ubicación de los archivos csv limpios
datos = os.listdir(ruta_archivo) #Se crea una variable que contiene una lista con los nombres de los archivos limpios, utlilzando la funcion os.listdir que entrega los nombres de los archivos de una carpeta.
lista = [] #Se crea una lista vacia con nombre "lista"

for i in range(0,len(datos)): # Se hace un ciclo for para recorrer la carpeta datos
  lista.append(pd.DataFrame(pd.read_csv(ruta_archivo + datos[i]))) #Utilizando la función .append en la variable "lista", se procede a agregar el contenido de los archivos csv.
#en cada iteración la función pd.read_csv, lee un archivo indicado por la ruta más el "i" de "datos", que corresponde al nombre del archivo.
pd.concat(lista).to_csv(ruta_archivo + "Base_Datos_Final.csv") #Se aplica la función pd.concat por sobre la lista con los datos y con la función ".to_csv" se guarda el archivo con los datos ya reunidos en la caréta resultados con el nombre "Base_Datos_Final.csv"


```
